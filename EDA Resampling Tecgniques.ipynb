```python
data = data.reset_index()
```

```python
data.info()
```

```python
data.set_index('Date', inplace = True)
data.head()
```

## Time Resampling
#### Rule A = Year End

```python
data.resample(rule ='A').min()
```

```python
data.resample(rule ='A').max()
```

```python
data.resample(rule ='A').max()['Open'].plot()
```

#### Rule - QS Quarterly

```python
data.resample(rule ='QS').max()
```

#### Rule - BA Buisness year end

```python
data.resample(rule ='BA').mean()
```

```python
 data.resample(rule ='BA').mean().plot(kind='bar', figsize=(12,4))
```

### Rolling

```python
data['Open'].rolling(10).mean()
```

```python
data['Open:10 days rolling'] = data['Open'].rolling(10).mean()
```

```python
data.head(11)
```

```python
data[['Open', 'Open:10 days rolling']].plot(figsize=(12,5))
```

```python
min_open_value = data[data['Open'] == data['Open'].min()]
print(min_open_value)
```
