```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
import yfinance as yf
```

```python
data = yf.download('TSLA')
```

```python
data.head()
```

```python
data.tail()
```

### Simple Moving Average(SMA)
#### Rolling() function

```python
data.Open.plot(figsize=(12,4))
```

```python
data['Oen 10 days rolling'] = data['Open'].rolling(window=10).mean()
data['Oen 30 days rolling'] = data['Open'].rolling(window=30).mean()
data['Oen 50 days rolling'] = data['Open'].rolling(window=50).mean()
```

```python
data.head(50)
```

```python
data[['Open', 'Oen 10 days rolling', 'Oen 30 days rolling', 'Oen 50 days rolling']].plot(xlim=['2020-01-01', '2020-12-01'], figsize = (12,4))
```

```python
### Advantages of SMA - Smoothining of the curve
### Disadvantages of SMA - Similar importance is given to all the data
```

### Cummulative Moving Average(CMA)
#### expanding() function

```python
data['CMU Open']=data['Open'].expanding().mean()
```

```python
data[['Open', 'CMU Open']].plot(xlim=['2020-01-01','2020-09-01'], figsize=(12,4))
```

### Exponential Moving Average(EMA)

```python
data['EMA 0.1'] = data['Open'].ewm(alpha = 0.1, adjust=False).mean()
data['EMA 0.3'] = data['Open'].ewm(alpha = 0.3, adjust=False).mean()
```

```python
data[['Open', 'EMA 0.1', 'EMA 0.3']].plot(xlim= ['2020-01-01', '2020-09-01'], figsize=(12,4))
```

```python
data.head()
```

### Exponential Weighted Moving Average (EWMA)

```python
data['EWMA 5days'] = data['Open'].ewm(span=5).mean()
```

```python
data.head()
```

```python
data[['Open', 'EWMA 5days', 'EMA 0.3']].plot(xlim=['2020-01-01', '2020-09-01'], figsize=(12,4))
```
